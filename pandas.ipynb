```python
import pandas as pd
import os
```

```python
def make_folders(filename):
    folders = os.listdir(filename)
    folders.remove(".DS_Store")
    return folders
```

```python
def make_fnames(folder):
    fnames = os.listdir("/Users/masontaylor/Downloads/Fall 2023/CSC-475/Project/Dataset/Clips/" + folder + "/")
    if ".DS_Store" in fnames:
        fnames.remove(".DS_Store")
    return fnames
    
```

```python
def ratings(fnames):
    ratings = []
    for file in fnames:
        i = file.find(",")
        rating = file[i-1:i+2]
        rating = rating.replace(",", ".")
        rating = float(rating)
        ratings.append(rating)
    return ratings
```

```python
def stance(fnames):
    stance = []
    for file in fnames:
        stance.append(file[0:1])
    return stance
```

```python
def tricks(fnames):
    tricks = []
    for file in fnames:
        tricks.append(file[2:file.find(",")-2])
    return tricks
```

```python
def split_tricks(fnames):
    trick1 = []
    trick2 = []
    trick3 = []
    for trick in tricks(fnames):
        split_tricks = trick.split(" + ")
        trick1.append(split_tricks[0])
        try:
            trick2.append(split_tricks[1])
        except:
            trick2.append("N/A")
        try:
            trick3.append(split_tricks[2])
        except:
            trick3.append("N/A")
    return trick1, trick2, trick3
```

```python
def make_dataframe(fnames):
    data = pd.DataFrame(fnames)
    data['rating'] = ratings(fnames)
    data['stance'] = stance(fnames)
    data['tricks'] = tricks(fnames)
    data.columns = ['video', 'rating', 'stance', 'tricks']
    data['trick1'], data['trick2'], data['trick3'] = split_tricks(fnames)
    return data
```

```python
def binarize_rating(data):
    binarized['binarized_rating'] = data['rating'].apply(lambda x: 0 if x == 0.0 else 1)
    return binarized
```

```python
def all_clips(filename):
    all_clips = pd.DataFrame()
    folders = make_folders(filename)
    for folder in folders:
        fnames = make_fnames(folder)
        clips = make_dataframe(fnames)
        #print(clips)
        info = folder.split(" ")
        clips['location'] = info[0]
        clips['year'] = int(info[1])
        all_clips = pd.concat([all_clips, clips])
    return all_clips
```
